## Visualisation

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.palettes import viridis
from bokeh.models import ColumnDataSource, LabelSet, Legend, FactorRange
from bokeh.transform import factor_cmap
from bokeh.layouts import gridplot
import numpy as np
import statsmodels.api as sm
```

```python
output_notebook()
```

```python
X = data["race"]
Y = data["sex"]
x_cats = sorted(X.unique())
y_cats = sorted(Y.unique())
sum([[data[(X == x_cat)&(Y == y_cat)].shape[0] for y_cat in y_cats] for x_cat in x_cats], [])
```

```python
def counts_by(X, Z, data=None, output_file_name=None):
    
    x_cats = sorted(Z.unique())
    y_cats = sorted(X.unique())
    
    x = [(x_cat, y_cat) for x_cat in x_cats for y_cat in y_cats]
    counts = sum(
        [[data[(Z == x_cat)&(X == y_cat)].shape[0] for y_cat in y_cats] for x_cat in x_cats], 
        []
    )
    
    source = ColumnDataSource(data=dict(
        x=x, 
        counts=counts
    ))

    p = figure(x_range=FactorRange(*x), plot_height=250, title="Fruit Counts by Year")

    p.vbar(
        x='x', top='counts', width=0.9, source=source,
        fill_color=factor_cmap('x', palette=viridis(len(y_cats)), factors=y_cats, start=1, end=2)
    )

    p.y_range.start = 0
    p.x_range.range_padding = 0.1
#     p.xaxis.major_label_orientation = 1
    p.xgrid.grid_line_color = None

    show(p)
    
counts_by(data["race"], data["sex"], data)
```

```python
from bokeh.models import FactorRange

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
years = ['2015', '2016', '2017']

data2 = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4],
        '2016'   : [5, 3, 3, 2, 4, 6],
        '2017'   : [3, 2, 4, 4, 5, 3]}

# this creates [ ("Apples", "2015"), ("Apples", "2016"), ("Apples", "2017"), ("Pears", "2015), ... ]
x = [ (fruit, year) for fruit in fruits for year in years ]
counts = sum(zip(data2['2015'], data2['2016'], data2['2017']), ()) # like an hstack

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), plot_height=250, title="Fruit Counts by Year")

p.vbar(x='x', top='counts', width=0.9, source=source)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```

```python
from bokeh.transform import factor_cmap

p = figure(x_range=FactorRange(*x), plot_height=250, title="Fruit Counts by Year")

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="white",

       # use the palette to colormap based on the the x[1:2] values
       fill_color=factor_cmap('x', palette=['firebrick', 'olive', 'navy'], factors=years, start=1, end=2))

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```
